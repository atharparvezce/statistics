# Statistics 

Statistics is the science of collecting , organising and analyzing data.

We are doing this for better decision making.

## Types of Statistics:-


### 1. Descriptive Statistics
It consists of organising and summarizing data.

__e.g.__ 

What is the average marks of the students in the class.

### 2. Inferential Statistics
Technique where in we used the data we have measured to form conclusions.

__e.g.__

Are the marks of the students of this classroom similar to the marks of the Maths classroom in the college ?

# Population and Sample 

## Population(N)
The entire group of individuals or items that you want to study.

**e.g.**

__a.__ All customers of a bank.

__b.__ Maths classrooms.

## Sample(n) 

A smaller group selected from the population for analysis.

**e.g.**  

__a.__ 1,000 randomly selected customers from the bank.

__b.__ Only one particularly selected class.

# Types of Sampling 


### 1. Simple Random Sampling
Every individual in the population has an equal chance of being selected.

**Example**

Picking 50 students randomly from a school.



### 2. Stratified Sampling
The population is divided into non overlapping groups(strata), and samples are taken from each group.

**Example**

Selecting students from each grade level in a school.



### 3. Systematic Sampling
Every nth individual from a list is selected.

**Example**

Choosing every 5th customer from a store’s entry list.



### 4. Convenience Sampling:
Samples are chosen based on ease of access.

**Example**

Surveying people at a nearby mall.

# Variables

Characteristics or properties that can change or take different values.

### 1. Quantitative Variables
Numerical values that can be measured or counted

**Example**

A person's height in centimeters.

#### 1.a. Discrete Variables

Countable numbers with no fractions or decimals.

**Example**

Number of students in a class.

#### 1.b. Continuous Variables
Measurable values that can have decimals.

**Example**

A person's weight in kilograms

### 2. Qualitative Variables
Descriptive categories or labels without numerical meaning.

**Example**

A person’s eye color (blue, brown, green).



# Variable Measurement Scale


### 1. Nominal Scale 
Categorizes data without any order.

**Example**

Types of fruits (apple, banana, mango).

### 2. Ordinal Scale
Categorizes data with a meaningful order but no fixed difference.

**Example**

Movie ratings (poor, average, good, excellent).

### 3. Interval Scale
Numeric scale with equal intervals but no true zero.

**Example**

Temperature in Celsius (0°C doesn’t mean no temperature).

### 4. Ratio Scale
Numeric scale with equal intervals and a true zero.

**Example**

Height in centimeters (0 cm means no height).


# Frequancy Distribution 
A table or graph that shows how often values occur in a dataset.

### 1. Bar Chart 
Uses bars to represent categorical data with gaps between bars.

**Example**

Number of students in different school clubs.


### 2. Histogram 
Uses bars to represent numerical data with no gaps, showing frequency distribution.

**Example**

Distribution of student heights in a class.

## Difference 
**Bar charts apply to discrete values, and histograms apply to continuous values.**

# Central Tendency
A measure that represents the center or typical value of a dataset.

### 1. Mean
The average of all values in a dataset. Mean does not work well with outlier.

**Example**

The average age of students in a class.

### 2. Median
The middle value when data is arranged in order. Median works well with outlier.

**Example** 

The median salary of employees in a company.

### 3. Mode 
The most frequently occurring value in a dataset.

**Example**

The most common shoe size in a store.

# Measure of Dispersion
Shows how spread out the data values are.

### Variance (σ² or s²)
Measures how far each value is from the mean.

**Example**

Variance helps determine how students' test scores differ from the average.

### Standard Deviation (σ or s)

**Example**

A lower standard deviation in exam scores means students performed consistently.

# Five-Number Summary 

The **Five-Number Summary** provides a quick overview of the distribution of data using five key values.

#### 1. **Minimum**
- The smallest value in the dataset.
- **Example:** For the dataset {10, 20, 30, 40, 50, 100}, the minimum is **10**.

#### 2. **First Quartile (Q1)**
- The **median of the lower half** of the data (25th percentile) OR
- The median of the values that are less than **Median(Q2)** .
- **Example:** In {10, 20, 30, 40, 50, 100}, divide the data into two halves:
  - Lower half: {10, 20, 30}
  - The median of the lower half is **20**, so **Q1 = 20**.

#### 3. **Median (Q2)**
- The middle value of the data (50th percentile).
- If the dataset has an odd number of elements, the median is the middle element. For an even number of elements, the median is the average of the two middle values.
- **Example:** In {10, 20, 30, 40, 50, 100}, the middle values are 30 and 40. So, the **median = (30 + 40) / 2 = 35**.

#### 4. **Third Quartile (Q3)**
- The **median of the upper half** of the data (75th percentile) OR
- The median of the values that are greater than **Median(Q2)** .
- **Example:** In {10, 20, 30, 40, 50, 100}, divide the upper half:
  - Upper half: {40, 50, 100}
  - The median of the upper half is **50**, so **Q3 = 50**.

#### 5. **Maximum**
- The largest value in the dataset.
- **Example:** In {10, 20, 30, 40, 50, 100}, the maximum is **100**.

---

### **Lower Fence and Higher Fence**
The **Lower Fence** and **Higher Fence** are used to detect outliers in the dataset. These fences are calculated using the **Interquartile Range (IQR)**.

**Formulae:**
- **IQR (Interquartile Range)** = \( Q3 - Q1 \)
- **Lower Fence** = \( Q1 - 1.5 \times IQR \)
- **Higher Fence** = \( Q3 + 1.5 \times IQR \)

#### **Step-by-Step Calculation:**

1. **Calculate the IQR:**
   IQR = Q3 - Q1 = 50 - 20 = 30
   

2. **Calculate the Lower Fence:**
   Lower Fence =20 - 45 = -25
   

3. **Calculate the Higher Fence:**
   Higher Fence =50 + 45 = 95

---

### **Summary:**

For the dataset {10, 20, 30, 40, 50, 100}, the **Five-Number Summary** is:

- **Minimum = 10**
- **Q1 = 20**
- **Median = 35**
- **Q3 = 50**
- **Maximum = 100**

The **Lower Fence** = **-25**, and the **Higher Fence** = **95**.

---

### **Outliers:**
- Any data point below the **Lower Fence** or above the **Higher Fence** is considered an outlier.
- In this dataset, **100** is an outlier because it is not in the range from **-25 to 95**.


```python
# Data
data = [10, 20, 30, 40, 50,100]

# Calculate the Five-Number Summary
minimum = np.min(data)
Q1 = np.percentile(data, 25)  # First Quartile (25th percentile)
median = np.median(data)      # Median (50th percentile)
Q3 = np.percentile(data, 75)  # Third Quartile (75th percentile)
maximum = np.max(data)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Calculate Lower and Higher Fences
lower_fence = Q1 - 1.5 * IQR
higher_fence = Q3 + 1.5 * IQR

# Print the results
print(f"Minimum: {minimum}")
print(f"Q1 (First Quartile): {Q1}")
print(f"Median (Q2): {median}")
print(f"Q3 (Third Quartile): {Q3}")
print(f"Maximum: {maximum}")
print(f"Interquartile Range (IQR): {IQR}")
print(f"Lower Fence: {lower_fence}")
print(f"Higher Fence: {higher_fence}")

# Outlier Check
outliers = [x for x in data if x < lower_fence or x > higher_fence]
print(f"Outliers: {outliers}")
```

**Note**:
- **Manual Method**: Picks the middle value directly (e.g., Q1 = 20).
- **Python Method**: Uses interpolation, averaging the middle values when there are even numbers, resulting in Q1 = 22.5.


# Distribution



## Gaussian (Normal) Distribution

A Gaussian distribution, also called a normal distribution, is a bell-shaped curve where most values cluster around the mean, and fewer values appear as you move further from the mean.

**Example**

Heights of people in a population follow a normal distribution, where most people have an average height, and very few are extremely short or tall.

## Standard Normal Distribution

A standard normal distribution is a special case of the normal distribution where:

Mean (μ) = 0
Standard Deviation (σ) = 1

**Example**

If you convert students' exam scores into z-scores (standardized scores), the new distribution follows a standard normal distribution.

## Z-Score

A Z-score tells how many standard deviations a value is from the mean. It helps compare different data points in a standard way.

**Example**
If a student scores 85 on a test with a mean of 70 and a standard deviation of 10, their Z-score is:

**Z = (x-mean)/sd = (85-70)/10 = 1.5**

This means the score is 1.5 standard deviations above the mean.

**Z-score is used to convert a normal distribution into a standard normal distribution!**

# Python Code for above some concepts

```python
# Step 1: Import necessary libraries
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Step 2: Load the dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Step 3: Calculate Mean, Median, and Mode for total_bill
mean_value = np.mean(tips["total_bill"])
median_value = np.median(tips["total_bill"])
mode_value = stats.mode(tips["total_bill"], keepdims=True)[0][0]
```

```python
# Step 4: Print the values
print(f"Mean of total_bill: {mean_value}")
print(f"Median of total_bill: {median_value}")
print(f"Mode of total_bill: {mode_value}")
```

```python
# Step 5: Draw a box plot for total_bill
plt.figure(figsize=(6, 4))
sns.boxplot(x=tips["total_bill"])
plt.title("Box Plot of Total Bill")
plt.xlabel("Total Bill")
plt.show()
```

```python
# Step 6: Draw a Histogram for total_bill
plt.figure(figsize=(6, 4))
sns.histplot(tips["total_bill"], bins=20, kde=True, color="skyblue")
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```
